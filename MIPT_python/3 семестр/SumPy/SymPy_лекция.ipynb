# Лекция по sympy + scipy

```python
from sympy import symbols
```

```python
x, y = symbols('x y')
print(x)
```

```python
x, y = symbols('x y')

a = x + y + x
print(a)

a -= x
print(a)

a += (5 * y + 12)
print(a)
```

```python
x = symbols('x')
f = x ** 2 + 5 * x + 1
print(f)
print(f.evalf(subs={x: 2}))

print((x + x + x).evalf(subs={x : 1}))
```

```python
from sympy import Eq, solve

x = symbols('x')
f = x ** 2 + 5 * x + 1
e = Eq(f, 1)
print(e)
print(solve(e))
print('=============')
a = symbols('a')
f2 = x ** 2 + a * a * x
e2 = Eq(f2, 0)

print(e2)
print(solve(e2))
print(solve(e2, x))
```

# SciPy

```python
from scipy import linalg
import numpy as np
```

```python
a = np.array([
    [3, 2, 0],
    [1, -1, 0],
    [0, 5, 1]])
b = np.array([2, 4, -1])
x = linalg.solve(a, b)

print(x)
print(a @ x - b)
```

```python
import matplotlib.pyplot as plt
from scipy import fft
N = 600
T = 1.0 / 800.0

x = np.linspace(0.0, N * T, N)
y = np.sin(50.0 * 2.0 * np.pi * x) + 0.5 * np.sin(80.0 * 2.0 * np.pi * x)  + np.random.random((N, 1)).flatten()
plt.plot(x, y)
plt.show()

yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0 * T), N // 2)
plt.plot(xf, 2.0 / N * np.abs(yf[0:N//2]))
plt.show()
```
