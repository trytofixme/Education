```python
from scipy.optimize import minimize
from matplotlib import pylab as plt
import numpy as np


def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)


x = np.arange(1, 31, 1)

answer1 = round(minimize(f, 2.0, method = 'BFGS').fun, 2)
answer2 = round(minimize(f, 30.0, method = 'BFGS').fun, 2)
```

```python
x1 = np.arange(1, 31, 1)
y1 = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
plt.plot(x1, y1)
plt.show()
```

```python
q = open('submission-1.txt', 'w')
q.write(str(answer1) + ' ' + str(answer2))
q.close()
```
